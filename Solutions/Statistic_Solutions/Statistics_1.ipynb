## Objectives

### For 'House of Cards', what is the Show Id and Who is the Director of this show ?

```python
use_cols = ['title', 'show_id', 'director']
netflix_df[ (netflix_df['title'] == 'House of Cards')][use_cols]
```

### In which year were the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
col = 'release_year'
top_index = 5
frequency_value_counts = netflix_df[col].value_counts(normalize=True)*100
frequency_value_counts.index = [str(year) for year in frequency_value_counts.index]
frequency_value_counts
```

```python
plt.figure()
plt.gcf().set_size_inches(10, 5)
plt.bar(
    x=frequency_value_counts.index[:3], height=frequency_value_counts.to_numpy()[:3], color='tab:cyan'
)
plt.grid(False)
plt.title('Top 3 Producing Years')
plt.show()
```

### How many Movies & TV Shows are in the dataset ? Show with Bar Graph.

```python
col = 'type'
frequency_value_counts = netflix_df[col].value_counts(normalize=True)*100
frequency, destination = frequency_value_counts.to_numpy(), frequency_value_counts.index.to_numpy()
frequency_value_counts
```

```python
plt.figure()
plt.gcf().set_size_inches(4, 4)
plt.bar(x=destination, height=frequency, color='tab:green')
plt.grid(False)
plt.title('Movies and TV Shows')
plt.show()
```

```python
plt.figure()
plt.gcf().set_size_inches(4, 4)
plt.barh(y=destination, width=frequency, color='tab:green')
plt.grid(False)
plt.title('Movies and TV Shows')
plt.show()
```

### Show all the Movies that were released in year 2000.

### Show only the Titles of all TV Shows that were released in India only.

### Show Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix ?
    Using Bar Chart

```python
col = 'director'
director_value_counts = netflix_df[col].value_counts(normalize=True)*100
director_value_counts
```

```python
frequency, destination = director_value_counts.to_numpy()[:10], director_value_counts.index.to_numpy()[:10]
```

```python
plt.figure()
plt.gcf().set_size_inches(14, 10)
plt.barh(y=destination, width=frequency, color='tab:blue')
plt.grid(False)
plt.title('Top 10 Directors')
plt.gca().invert_yaxis()
plt.show()
```

### Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom".

```python
netflix_df['listed_in'] = netflix_df['listed_in'].apply(lambda entry: entry.split(','))
netflix_df['comedy_flag'] = netflix_df['listed_in'].apply(lambda entry: 'Comedies' in entry)
netflix_df.sample(3)
```

```python
cols_to_use = ['title', 'type', 'listed_in', 'country', 'comedy_flag']
netflix_df[((netflix_df['type'] == 'Movie') & (netflix_df['comedy_flag'] == True)) & (netflix_df['country'] == 'United Kingdom')][cols_to_use]
```

### In how many movies/shows, Tom Cruise was cast ?

```python
netflix_df['tc_flag'] = netflix_df['cast'].apply(lambda entry: 'Tom Cruise' in entry)
netflix_df.sample(3)
```

```python
cols_to_use = ['title', 'tc_flag']
netflix_df[((netflix_df['tc_flag'] == True))][cols_to_use]
```

### How many Movies got the 'TV-14' rating, in Canada ?

```python
cols_to_use = ['title', 'type', 'rating', 'country']
netflix_df[((netflix_df['type'] == 'Movie') & (netflix_df['rating'] == 'TV-14')) & (netflix_df['country'] == 'Canada')][cols_to_use]
```

### How many TV Shows got the 'R' rating, after year 2018 ?

```python
netflix_df[
    (netflix_df['release_year'] > 2018) & (netflix_df['rating'] == "R") & (netflix_df['type'] == 'Movie')
][cols_to_use]
```

### What is the maximum duration of a Movie/Show on Netflix ?

```python
netflix_df['duration'] = netflix_df['duration'].fillna('0 min') #Filling all values with NaN with '0 minutes'
```

```python
# Maximum duration for movies
max_movie_dur = netflix_df[
    (netflix_df['type'] == 'Movie')
]['duration'].apply(lambda entry: int(entry.split()[0])).max()
print(f"{max_movie_dur} min")
```

```python
# Maximum duration for movies
max_tvshow_dur = netflix_df[
    (netflix_df['type'] == 'TV Show')
]['duration'].apply(lambda entry: int(entry.split()[0])).max()
print(f"{max_tvshow_dur} seasons")
```

### Find all the instances where: Category is 'Movie' and Type is 'Dramas' or Category is 'TV Show' & Type is 'Kids' TV'.

```python
netflix_df['kids_flag'] = netflix_df['listed_in'].apply(lambda entry: 'Kids TV' in entry)
netflix_df['drama_flag'] = netflix_df['listed_in'].apply(lambda entry: 'Dramas' in entry)
netflix_df.sample(3)
```

```python
cols_to_use = ['title', 'type', 'kids_flag', 'drama_flag']
netflix_df[((netflix_df['type'] == 'Movie') & (netflix_df['drama_flag'] == True)) | ((netflix_df['type'] == 'Tv Show') & netflix_df['kids_flag'] == True)][cols_to_use]
```
